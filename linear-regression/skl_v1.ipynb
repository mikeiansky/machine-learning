```python
import numpy as np
import matplotlib.pyplot as plt

dataset = np.genfromtxt('train.csv', delimiter=',')

dataset.shape
```

```python
dataset = dataset[1:,:]
```

```python
fig, ax = plt.subplots()
ax.scatter(dataset[:,0],dataset[:,1]) 

plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(dataset[:,0].reshape(-1,1),dataset[:,1].reshape(-1,1))
```

```python
# 从训练好的模型中提取系数和截距
w = lr.coef_[0][0]
b = lr.intercept_[0]

print("w is: ", w)
print("b is: ", b)
```

```python
def cost(w, b):
    predict_y = dataset[:,0] * w + b
    return np.sum((dataset[:,1] - predict_y) ** 2)/ dataset[:,0].shape[0] 
```

```python
print('cost', cost(w,b))
```

```python
plt.scatter(dataset[:,0],dataset[:,1])


x_vals = np.array(ax.get_xlim())
y_vals = b + w * x_vals
plt.plot(x_vals, y_vals, color='red')

plt.show()
```
